```python
class Mecanico:
    def __init__(self, nombre, telefono, edad):
        # Atributos de la clase
        self.nombre = nombre
        self.telefono = telefono
        self.edad = edad
    # Función 1: Diagnosticar
    def diagnosticar(self, vehiculo):
        print(f"🔧 {self.nombre} está diagnosticando el motor del {vehiculo}...")
    # Función 2: Reparar
    def reparar(self, pieza):
        print(f"🛠️ {self.nombre} está reparando: {pieza}. ¡Quedará como nuevo!")
    # Función 3: Cambio de aceite
    def cambio_aceite(self):
        print(f"🛢️ {self.nombre} está realizando el cambio de aceite y filtro.")
# --- Creación de Objetos (Instancias) ---
# Objeto 1: mecanico1
mecanico1 = Mecanico("Juan Pérez", "555-0123", 35)
# Objeto 2: mecanico2
mecanico2 = Mecanico("Ana Martínez", "555-9876", 28)
# --- Pruebas de funcionamiento ---
print(f"--- Ficha del Mecánico 1: {mecanico1.nombre} ---")
mecanico1.diagnosticar("Sedán azul")
mecanico1.reparar("los frenos")
mecanico1.cambio_aceite()
mecanico1.telefono = "555-5555"
print(f"--- Ficha del Mecánico 2: {mecanico2.nombre} ---")
mecanico2.cambio_aceite()
mecanico2.diagnosticar("SUV rojo")
mecanico2.reparar("el motor")
print(f"Contacto: {mecanico2.telefono}")

print("Programa realizado por Meredith Agurre NL = 2")
```
